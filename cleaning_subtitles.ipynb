```python
# taking 20% data randomly

df1 = df.sample(frac = 0.2,random_state = 42)
df1
```

```python
# function to cleaning the subtitle text

import re

def clean_text(text):
    text = re.sub(r'\d+\s+\d{1,2}:\d{2}:\d{2},\d{3} --> \d{1,2}:\d{2}:\d{2},\d{3}\r\n', '', text)
    text = re.sub(r'\r\n', ' ', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    text = text.lower()
    return text
```

```python
#cleaning the subtitle text
df1['subtitles'] = df1['file_content'].apply(clean_text)
```

```python
#dropping unnecessary columns
df2 = df1.drop(['content','file_content'],axis =1)
```

```python
# making multiple csv files
import pandas as pd
import os

# Sample DataFrame
df = df2

# Define the number of rows per file
rows_per_file = 3000  # Adjust based on your needs

# Calculate the number of smaller files needed
num_files = len(df) // rows_per_file + 1

# Directory to save the smaller files
output_directory = 'C:\\Users\\MSI\\Desktop\\clone_project\\subtitles_csv'

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Split and save the DataFrame into multiple files
for i in range(num_files):
    start_idx = i * rows_per_file
    end_idx = min((i + 1) * rows_per_file, len(df))
    
    # Get the chunk of the DataFrame
    chunk = df.iloc[start_idx:end_idx]
    
    # Define the output file name
    output_file = os.path.join(output_directory, f"subtitle_{i+1}.csv")
    
    # Save the chunk as a new CSV file
    chunk.to_csv(output_file, index=False)

print(f"DataFrame has been split into {num_files} smaller files.")
```
