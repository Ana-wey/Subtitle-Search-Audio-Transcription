```python
import sys
import csv

from langchain_community.document_loaders.csv_loader import CSVLoader

from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_community.embeddings import SentenceTransformerEmbeddings
from sentence_transformers import SentenceTransformer

from langchain_chroma import Chroma
from langchain.docstore.document import Document

import chromadb
#from chromadb.api.types import Document
from chromadb.utils import embedding_functions
#from chromadb import Document
```

```python
import pandas as pd
import os

# Directory where your CSV files are stored
directory_path = '/content/drive/MyDrive/shazam_project'

# List all files in the directory
csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]

# Read each CSV file and store them in a list
for file in csv_files:
    file_path = os.path.join(directory_path, file)
    df = pd.read_csv(file_path)

    # Increase the field size limit
    csv.field_size_limit(2147483647)
    # Load the CSV file using the CSVLoader

    loader = CSVLoader(file_path=file_path,   # Path to the current CSV file
                       source_column='name',  # Column for document names
                       metadata_columns=['num','name'], # Metadata columns to extract
                       content_columns=['subtitles'] # Column for text content
                       )
    data = loader.load()  # Load data as Document objects


    # Extract contents and metadata from the documents
    doc_contents = [doc.page_content for doc in data]
    meta_datas = [doc.metadata for doc in data]

    # Initialize text splitter
    text_splitter = RecursiveCharacterTextSplitter(
        separators=["\n\n", "\n", " ", ""],
        chunk_size=500,
        chunk_overlap=50,
    )

    # Split the content into smaller chunks
    chunks = text_splitter.create_documents(doc_contents, meta_datas)


    # initialising the embedding model
    model_name = "all-MiniLM-L6-v2"
    embeddings_model = SentenceTransformer(model_name)


    # Create the PersistentClient
    client = chromadb.PersistentClient(path="/content/drive/MyDrive/shazam_project/chroma_db")

    # Create the embedding function (assuming embeddings_model is already defined)
    embedding_function = embedding_functions.SentenceTransformerEmbeddingFunction(model_name=model_name)


    # Create or retrieve the collection from the client
    collection = client.get_or_create_collection(name="subtitle_database", embedding_function=embedding_function)

    # Prepare documents (chunks) as a list of Document objects
    documents = [
        Document(page_content=doc.page_content, metadata=doc.metadata)
        for doc in chunks
    ]

    batch_size = 500  # Set your desired batch size (below the limit)
    num_batches = (len(documents) + batch_size - 1) // batch_size

    # Add documents in batches
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(documents))
        batch_documents = documents[start_idx:end_idx]

        collection.add(
            documents=[doc.page_content for doc in batch_documents],
            metadatas=[doc.metadata for doc in batch_documents],
            ids=[str(idx) for idx in range(start_idx, end_idx)]
        )
```
