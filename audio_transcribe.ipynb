```python

import whisper

# Load the Whisper model (you can use other models like "small", "medium", etc.)
model = whisper.load_model("base")

def transcribe_audio(audio_file):
    # Transcribe the audio file
    result = model.transcribe(audio_file)
     # Get the transcribed text from the result
    transcribed_text = result["text"]
    return transcribed_text

# Path to the audio file
audio_file = "/content/drive/MyDrive/WhatsAppPM.mpeg"

# Transcribe the audio and print the result
transcribed_text = transcribe_audio(audio_file)

print(transcribed_text)
```

```python
# Your query text
query = transcribed_text

# Initialize the embedding model (same one used for documents)
from sentence_transformers import SentenceTransformer

model_name = "all-MiniLM-L6-v2"
embedding_model = SentenceTransformer(model_name)

# Generate the embedding for the query
query_embedding = embedding_model.encode([query])[0]  # Encode the query into a vector
```

```python
import chromadb
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')

# Initialize the Chroma PersistentClient (same database path)
client = chromadb.PersistentClient(path="/content/drive/MyDrive/Colab_Notebooks/chroma_db")

# Retrieve the collection
collection = client.get_or_create_collection(name="vector_database")

# Perform the similarity search by querying the collection with the query embedding
# We retrieve the top 5 most similar documents
results = collection.query(
    query_embeddings=[query_embedding],  # The query embedding
    n_results=5  # Number of similar results to retrieve
)

# Display the results
for i, metadata in enumerate(results["metadatas"][0]):
        subtitle_name = metadata["name"]
        subtitle_id = metadata["num"]
        url = f"https://www.opensubtitles.org/en/subtitles/{subtitle_id}"
        print(f"Result {i + 1}:")
        print(f"  Subtitle Name: {subtitle_name.upper()}")
        print(f"  URL: {url}")
        print("-" * 20)
```
