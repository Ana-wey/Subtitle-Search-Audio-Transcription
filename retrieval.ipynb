```python
# Your query text
query ="man davon wheres the boy groaning gasping wheezing groans grunting grunts gasps je verrai grunting gasps grunts groans indistinct shouting in distance indistinct shouting oh my head groans help coughs over here indistinct shouting gasps woman murderer child citron speaking indistinctly gasps please get help mama davon speaking french wait wait girl im not gonna hurt you murderer "

# Initialize the embedding model (same one used for documents)
from sentence_transformers import SentenceTransformer

model_name = "all-MiniLM-L6-v2"
embedding_model = SentenceTransformer(model_name)

# Generate the embedding for the query
query_embedding = embedding_model.encode([query])[0]  # Encode the query into a vector
```

```python
import chromadb
import warnings

# Ignore all warnings
warnings.filterwarnings('ignore')

# Initialize the Chroma PersistentClient (same database path)
client = chromadb.PersistentClient(path="/content/drive/MyDrive/shazam_project/chroma_db")

# Retrieve the collection
collection = client.get_or_create_collection(name="subtitle_database")

# Perform the similarity search by querying the collection with the query embedding
# We retrieve the top 5 most similar documents
results = collection.query(
    query_embeddings=[query_embedding],  # The query embedding
    n_results=5  # Number of similar results to retrieve
)

# Display the results
for i, metadata in enumerate(results["metadatas"][0]):
        subtitle_name = metadata["name"]
        subtitle_id = metadata["num"]
        url = f"https://www.opensubtitles.org/en/subtitles/{subtitle_id}"
        print(f"Result {i + 1}:")
        print(f"  Subtitle Name: {subtitle_name.upper()}")
        print(f"  URL: {url}")
        print("-" * 20)
```
